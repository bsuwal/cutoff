```julia
X₀_Dist = Uniform(-1, 1)

# WeightDist = Normal(0.0, 1)

# activation = σ

activation = tanh
num_chains = 20000

num_steps = 15
forward = true
store_steps = false

# plotting stuff
p = plot()

for N = 1:15
    grid_size = 0.001
    ϵ = grid_size/2
    println("N=$N")
    WeightDist = Uniform(-1/√N, 1/√N)
#     WeightDist = Normal(0.0, 1)
#     WeightDist = Normal(0.0, 1/√N)
#     WeightDist = Uniform(-1, 1)
    
    Exp = Experiment(X₀_Dist, N, num_chains, WeightDist, activation, grid_size, num_steps, forward, store_steps)
    Results = ExperimentResults([], [], [])
    
    run_chain(Exp, Results)
    
    diststr_greek, diststr_nongreek, actstr, dynamicsstr = get_plotting_strs(Exp)
    if activation == σ
        title = "$diststr_greek, $actstr, N=$N, \n $dynamicsstr, $num_chains chains"
    elseif activation == tanh
        title = "$diststr_greek, $actstr, N=$N, \n $dynamicsstr, $num_chains chains, ϵ=$ϵ"
    end
    plot!(Results.tvds,
         title=title,
         xlabel="# layers",
         ylabel="tvd",
         xlim=(0, num_steps),
         ylim=(-0.2, 1.2),
         yticks = 0:0.2:1.2,
         seriestype=:scatter, 
         label="N=$N"
    )
end
display(p)
```
